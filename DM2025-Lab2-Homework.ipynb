## 2. Feature Engineering Steps

### 2.1 Stop words removal

```python
### For emoji removal
import re

def remove_emojis(text: str) -> str:
    """
    Remove emojis from a given string using Unicode ranges.
    Covers emoticons, pictographs, transport symbols, flags, etc.
    """
    if not isinstance(text, str):
        raise TypeError("Input must be a string.")

    # Regex pattern for emojis (covers most Unicode emoji ranges)
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # Emoticons
        "\U0001F300-\U0001F5FF"  # Symbols & pictographs
        "\U0001F680-\U0001F6FF"  # Transport & map symbols
        "\U0001F1E0-\U0001F1FF"  # Flags
        "\U00002700-\U000027BF"  # Dingbats
        "\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
        "\U00002600-\U000026FF"  # Misc symbols
        "\U00002B00-\U00002BFF"  # Arrows
        "]+",
        flags=re.UNICODE
    )

    return emoji_pattern.sub('', text)
```

```python
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import nltk

nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)

# Get English stop words
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text_with_stopwords(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = text.split()
    # Lemmatize tokens and remove stop words
    filtered_tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return filtered_tokens

# Apply the updated preprocessing to both dataframes
train_df['text'] = train_df['text'].apply(remove_emojis)
train_df['preprocessed_text_stopwords'] = train_df['text'].apply(preprocess_text_with_stopwords)
test_df['text'] = test_df['text'].apply(remove_emojis)
test_df['preprocessed_text_stopwords'] = test_df['text'].apply(preprocess_text_with_stopwords)

print("Text preprocessing with lemmatization and stop word removal complete. Displaying first few rows of train_df:")
print(train_df[['text', 'preprocessed_text_stopwords']].head())
```

### 2.2 Bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.model_selection import train_test_split

### Deep Learning Model + N-gram
X_train, X_eval, y_train, y_eval = train_test_split(train_df.drop(['emotion'], axis=1),train_df['emotion'], test_size=0.2, random_state=42)

# Join the preprocessed tokens back into a string for CountVectorizer
train_text_for_vectorizer = [' '.join(tokens) for tokens in X_train['preprocessed_text_stopwords']]
eval_text_for_vectorizer = [' '.join(tokens) for tokens in X_eval['preprocessed_text_stopwords']]

# Initialize CountVectorizer to include unigrams and bigrams
BOW_vectorizer_ngram = CountVectorizer(max_features = 1000, min_df=2) # Increased max_features for n-grams

X_train_BOW_ngram = BOW_vectorizer_ngram.fit_transform(train_text_for_vectorizer)
X_eval_BOW_ngram = BOW_vectorizer_ngram.transform(eval_text_for_vectorizer)
```

### 2.3 One hot encoder

```python
from sklearn.preprocessing import LabelEncoder
import numpy as np
import keras

label_encoder = LabelEncoder()
label_encoder.fit(y_train)
label_encoder.fit(y_eval)

print('check label: ', label_encoder.classes_)
print('\n## Before convert')
print('y_train[0:4]:\n', y_train[0:4])
print('\ny_train.shape: ', y_train.shape)

def label_encode(le, labels):
    enc = le.transform(labels)
    return keras.utils.to_categorical(enc)

def label_decode(le, one_hot_label):
    dec = np.argmax(one_hot_label, axis=1)
    return le.inverse_transform(dec)

y_train = label_encode(label_encoder, y_train)
y_eval = label_encode(label_encoder, y_eval)

print('\n\n## After convert')
print('y_train[0:4]:\n', y_train[0:4])
print('\ny_train.shape: ', y_train.shape)
```

```python
# I/O check
input_shape = X_train_BOW_ngram.shape[1]
print('input_shape: ', input_shape)

output_shape = len(label_encoder.classes_)
print('output_shape: ', output_shape)
```

## 3. Model Implementation Steps

### 3.1 Build deep learning model
Loss function: Focal Loss

Optimizer: AdamW

```python
from keras.models import Model
from keras.layers import Input, Dense, RNN
from keras.layers import ReLU, Softmax, Dropout
from keras.optimizers import AdamW
import tensorflow as tf

def categorical_focal_loss(gamma=2.0, alpha=0.5):
    """
    Softmax focal loss.
    """
    def focal_loss_fixed(y_true, y_pred):
        # Scale predictions so that the class probabilites sum to 1
        y_pred /= tf.reduce_sum(y_pred, axis=-1, keepdims=True)

        # Clip the prediction value to prevent NaN's and Inf's
        epsilon = tf.keras.backend.epsilon()
        y_pred = tf.clip_by_value(y_pred, epsilon, 1. - epsilon)

        # Calculate cross entropy
        cross_entropy = -y_true * tf.math.log(y_pred)

        # Calculate weight factor
        weight = alpha * tf.pow(1 - y_pred, gamma)

        # Calculate focal loss
        loss = weight * cross_entropy

        # Sum the losses in mini_batch
        return tf.reduce_sum(loss, axis=-1)

    return focal_loss_fixed

# input layer
model_input = Input(shape=(input_shape, ))
X = model_input

# 1st hidden layer
X = Dense(units=144)(X)
X = ReLU()(X)
X = Dropout(0.6)(X)

# 2nd hidden layer
X = Dense(units=72)(X)
X = ReLU()(X)
X = Dropout(0.3)(X)

# 3rd hidden layer
X = Dense(units=36)(X)
X = ReLU()(X)

# 4rd hidden layer
X = Dense(units=24)(X)
X = ReLU()(X)

# Output layer
X = Dense(units=output_shape)(X)
model_output = Softmax()(X)

# create model
model = Model(inputs=[model_input], outputs=[model_output])

# Define the optimizer with a learning rate
adam_optimizer = AdamW(learning_rate=7e-4, clipvalue=1)

# loss function & optimizer
model.compile(optimizer=adam_optimizer,
              loss=categorical_focal_loss(gamma=2.0, alpha=0.5), # Use the custom focal loss function
              metrics=['accuracy'])

# show model construction
model.summary()
```

### 3.2 Training and Evaluation

```python
import tensorflow as tf
import numpy as np
from scipy.sparse import csr_matrix

from keras.callbacks import CSVLogger

csv_logger = CSVLogger('training_log.csv')

# training setting
epochs = 25
batch_size = 128

# training!
history = model.fit(X_train_BOW_ngram.toarray(), y_train,
                    epochs=epochs,
                    batch_size=batch_size,
                    callbacks=[csv_logger],
                    validation_data = (X_eval_BOW_ngram.toarray(), y_eval))
print('training finish')
```

```python
### Print out evaluation reports
from sklearn.metrics import accuracy_score, classification_report

y_pred_train = model.predict(X_train_BOW_ngram.toarray(), batch_size=128)
y_pred_eval = model.predict(X_eval_BOW_ngram.toarray(), batch_size=128)

y_train_decode = label_decode(label_encoder, y_train)
y_eval_decode = label_decode(label_encoder, y_eval)
y_pred_train = label_decode(label_encoder, y_pred_train)
y_pred_eval = label_decode(label_encoder, y_pred_eval)

acc_train = accuracy_score(y_true=y_train_decode, y_pred=y_pred_train)
acc_eval = accuracy_score(y_true=y_eval_decode, y_pred=y_pred_eval)

print(f'Train accuracy: {round(acc_train,2)}')
print(f'Evaluation Accuracy: {round(acc_eval,2)}')
print(classification_report(y_true=y_eval_decode, y_pred=y_pred_eval))
```
